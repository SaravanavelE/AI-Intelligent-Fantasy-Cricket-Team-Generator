```python
import pandas as pd
import requests
import datetime

# Load the dataset
file_path = "/content/drive/MyDrive/Gameathon/Gameathon-CricTech/Dataset/weather dataset for gameathon.csv"
df = pd.read_csv(file_path)

# OpenWeatherMap API Key
API_KEY = API_KEY

# Function to classify weather conditions and print the team composition ratio
def classify_weather(description):
    description = description.lower()
    weather_teams = {
        "sunny": "5:2:2:2",
        "cloudy": "3:3:3:2",
        "rainy": "4:2:3:2",
        "windy": "3:4:2:2",
        "foggy/misty": "2:4:3:2",
        "humid": "3:3:3:2",
        "cold/winter": "3:4:2:2",
        "autumn/fall": "3:3:3:2",
        "overcast": "2:4:3:2",
        "dewy": "4:2:3:2",
        "unknown": "3:4:2:2"
    }
    for key in weather_teams:
        if key in description:
            print(f"Recommended Team (Weather): {weather_teams[key]}")
            return key, weather_teams[key]
    print("Recommended Team (Weather): 3:4:2:2")
    return "unknown", "3:4:2:2"

# Function to classify pitch type and return team composition
def classify_pitch(pitch_type):
    pitch_teams = {
        "Batting-friendly": "5:2:2:2",
        "Spin-friendly": "3:4:2:2",
        "Balanced": "3:3:3:2",
        "Bowler-friendly": "2:4:3:2",
        "New venue": "3:3:3:2",
        "Slow, Spin-friendly": "3:4:2:2",
        "Seam-friendly": "3:3:3:2"
    }
    team = pitch_teams.get(pitch_type, "3:3:3:2")
    print(f"Recommended Team (Pitch): {team}")
    return team

# Function to determine the final combined recommendation
def get_final_team(weather_team, pitch_team):
    final_team = [max(int(w), int(p)) for w, p in zip(weather_team.split(":"), pitch_team.split(":"))]
    final_team_str = ":".join(map(str, final_team))
    print(f"Final Recommended Team: {final_team_str}")
    return final_team_str

# Function to get weather forecast using OpenWeatherMap API
def get_forecast(lat, lon, match_datetime):
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    data = response.json()
    if response.status_code == 200 and "list" in data:
        forecast_data = data["list"]
        best_match = min(forecast_data, key=lambda x: abs(datetime.datetime.strptime(x["dt_txt"], "%Y-%m-%d %H:%M:%S") - match_datetime))
        description = best_match["weather"][0]["description"]
        classified_condition, weather_team = classify_weather(description)
        return {
            "Temperature (°C)": best_match["main"]["temp"],
            "Weather Condition": description,
            "Classified Condition": classified_condition,
            "Weather Team": weather_team
        }
    else:
        return {"Error": "Forecast data not available"}

# Function to get weather conditions for a user-input match number
def get_match_weather():
    match_number = int(input("Enter Match Number: "))
    match = df[df["Match Number"] == match_number]
    if match.empty:
        return "Match number not found!"
    venue = match.iloc[0]["Venue"]
    home = match.iloc[0]["Team1"]
    away = match.iloc[0]["Team2"]
    lat, lon = match.iloc[0]["Latitude"], match.iloc[0]["Longitude"]
    pitch_type = match.iloc[0]["Pitch Type"]
    match_date = "2025-04-04" # match.iloc[0]["Date"]
    match_time = "15:30"  # match.iloc[0]["Time"]
    if pd.isnull(lat) or pd.isnull(lon):
        return f"Coordinates not available for {venue}"
    match_datetime = datetime.datetime.strptime(f"{match_date} {match_time}", "%Y-%m-%d %H:%M")
    weather = get_forecast(lat, lon, match_datetime)
    pitch_team = classify_pitch(pitch_type)
    final_team = get_final_team(weather.get("Weather Team", "3:3:3:2"), pitch_team)
    print("\nMatch Weather and Pitch Details:")
    print(f"Match Number: {match_number}")
    print(f"Venue: {venue}")
    print(f"Home: {home}")
    print(f"Away: {away}")
    print(f"Date: {match_date}")
    print(f"Time: {match_time}")
    print(f"Temperature: {weather.get('Temperature (°C)', 'N/A')}°C")
    print(f"Weather Condition: {weather.get('Weather Condition', 'N/A')}")
    print(f"Classified as: {weather.get('Classified Condition', 'N/A')}")
    print(f"Pitch Type: {pitch_type}")
    print(f"Final Team Composition: {final_team}")

# Run the function
get_match_weather()
```

```python
import pandas as pd
import requests
import datetime

# Load the dataset
file_path = "/content/drive/MyDrive/Gameathon/Gameathon-CricTech/Dataset/weather dataset for gameathon.csv"
df = pd.read_csv(file_path)

# OpenWeatherMap API Key
API_KEY = API_KEY

# Function to classify weather conditions
weather_team_config = {
    "sunny": [5, 2, 2, 2],
    "cloudy": [3, 3, 3, 2],
    "rainy": [4, 2, 3, 2],
    "windy": [3, 4, 2, 2],
    "foggy": [2, 4, 3, 2],
    "humid": [3, 3, 3, 2],
    "cold": [3, 4, 2, 2],
    "autumn": [3, 3, 3, 2],
    "overcast": [2, 4, 3, 2],
    "dewy": [4, 2, 3, 2],
    "unknown": [3, 4, 2, 2]
}

def classify_weather(description):
    description = description.lower()
    for key in weather_team_config.keys():
        if key in description:
            return key, weather_team_config[key]
    return "unknown", weather_team_config["unknown"]

# Function to classify pitch type
pitch_team_config = {
    "batting-friendly": [5, 2, 2, 2],
    "spin-friendly": [3, 3, 3, 2],
    "balanced": [4, 3, 2, 2],
    "bowler-friendly": [3, 4, 2, 2],
    "slow, spin-friendly": [3, 3, 3, 2],
    "seam-friendly": [2, 4, 3, 2],
    "new venue": [3, 3, 3, 2]
}

def classify_pitch(pitch_type):
    return pitch_team_config.get(pitch_type.lower(), [3, 4, 2, 2])

# Function to merge team configurations
def merge_team_configs(weather_config, pitch_config):
    combined = [(w + p) // 2 for w, p in zip(weather_config, pitch_config)]
    diff = 11 - sum(combined)
    combined[0] += diff  # Adjust batsmen count to ensure total is 11
    return combined

# Function to get weather forecast

def get_forecast(lat, lon, match_datetime):
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    data = response.json()

    if response.status_code == 200 and "list" in data:
        forecast_data = data["list"]
        best_match = min(forecast_data, key=lambda x: abs(datetime.datetime.strptime(x["dt_txt"], "%Y-%m-%d %H:%M:%S") - match_datetime))
        description = best_match["weather"][0]["description"]
        classified_condition, weather_config = classify_weather(description)

        return {
            "Temperature (°C)": best_match["main"]["temp"],
            "Weather Condition": description,
            "Classified Condition": classified_condition,
            "Weather Team": weather_config,
        }
    else:
        return {"Error": "Forecast data not available"}

# Function to get weather and pitch-based recommendations
def get_match_weather():
    match_number = int(input("Enter Match Number: "))
    match = df[df["Match Number"] == match_number]
    if match.empty:
        return "Match number not found!"

    venue = match.iloc[0]["Venue"]
    home = match.iloc[0]["Team1"]
    away = match.iloc[0]["Team2"]
    match_date = "2025-04-04" # match.iloc[0]["Date"]
    match_time = "15:30"  # match.iloc[0]["Time"]
    lat, lon = match.iloc[0]["Latitude"], match.iloc[0]["Longitude"]
    pitch_type = match.iloc[0]["Pitch Type"]
    match_date = "2025-04-03"
    match_time = "19:30"

    if pd.isnull(lat) or pd.isnull(lon):
        return f"Coordinates not available for {venue}"

    match_datetime = datetime.datetime.strptime(f"{match_date} {match_time}", "%Y-%m-%d %H:%M")
    weather = get_forecast(lat, lon, match_datetime)
    pitch_config = classify_pitch(pitch_type)

    if "Error" in weather:
        return weather["Error"]

    final_team_config = merge_team_configs(weather["Weather Team"], pitch_config)

    print("\nMatch Weather and Pitch-Based Recommendations:")
    print(f"Match Number: {match_number}")
    print(f"Venue: {venue}")
    print(f"Home: {home}")
    print(f"Away: {away}")
    print(f"Date: {match_date}")
    print(f"Time: {match_time}")
    print(f"Temperature: {weather['Temperature (°C)']}°C")
    print(f"Weather Condition: {weather['Weather Condition']}")
    print(f"Classified as: {weather['Classified Condition']}")
    print(f"Weather-Based Team: {weather['Weather Team']}")
    print(f"Pitch-Based Team: {pitch_config}")
    print(f"Final Recommended Team: {final_team_config} (Batsmen: {final_team_config[0]}, All-Rounders: {final_team_config[1]}, Bowlers: {final_team_config[2]}, Wicketkeepers: {final_team_config[3]})")

# Run the function
get_match_weather()
```
